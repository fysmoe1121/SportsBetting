```python
import json
file_path = '/Users/leonma/Downloads/projections.json'

def read_partial_json(file_path, n_bytes=1024):
    with open(file_path, 'r') as file:
        partial_content = file.read(n_bytes)
    try:
        # Attempt to parse the partial content as JSON
        partial_data = json.loads(partial_content)
        print("Partial JSON data:", partial_data)
    except json.JSONDecodeError:
        print("The partial content is not valid JSON. Showing raw content instead:")
        print(partial_content)

# Specify the path to your large JSON file
# Call the function, adjusting n_bytes as needed
read_partial_json(file_path, n_bytes=1024)
```

```python
print("Before loading the JSON")

try:
    with open(file_path, 'r') as file:
        data = json.load(file)
        print("Loaded JSON:", data)
except Exception as e:
    print("Error:", e)

print("After loading the JSON")
```

```python
with open(file_path, 'r') as file:
    data = json.load(file)

def safe_get(d, *keys):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key)
        else:
            return None
    return d

# Use the safe_get function to filter the projections
filtered_projections = [
    projection for projection in data
    if safe_get(projection, "relationships", "league", "data", "id") == "265"
]

# Print the filtered projections
print(json.dumps(filtered_projections, indent=2))
```
